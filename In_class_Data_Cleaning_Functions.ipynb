```python
def nba_wrangling(df):
  import pandas as pd

  # Drop 'Unnamed: 0' column only if it exists
  if 'Unnamed: 0' in df.columns:
    df.drop(columns=['Unnamed: 0'], inplace=True)

  # Replace 'PG-SG' with 'PG' in the 'Position' column
  df.loc[df['Position'] == 'PG-SG', 'Position'] = 'PG'
  df.loc[df['Position'] == 'SG-SF', 'Position'] = 'SG'
  df.loc[df['Position'] == 'SF-PF', 'Position'] = 'SF'
  df.loc[df['Position'] == 'PF-C', 'Position'] = 'PF'
  df.loc[df['Position'] == 'C-PF', 'Position'] = 'C'
  df.loc[df['Position'] == 'SF-SG', 'Position'] = 'SF'
  df.loc[df['Position'] == 'SG-PG', 'Position'] = 'SG'

  return df
```

```python
import functions as fn

# Begin my ML pipeline
df = pd.read_csv('./data/nba_salaries.csv')
df = nba_wrangling(df)
df = fn.basic_wrangling(df)
df = fn.manage_dates(df)
df = fn.bin_categories(df, min_records=10)
df.head(10)
# import sys
# sys.path.append('../../../functions')
```
